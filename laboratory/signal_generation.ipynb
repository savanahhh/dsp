1. Unit Sample Sequence

```python
#unit sample sequence

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11)

unit_sample = np.zeros(n.shape)
unit_sample[n == 0] = 1

plt.stem(n, unit_sample, use_line_collection=True)
plt.title('Unit Sample Sequence')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.5)
plt.show()
```

2. Unit Step Signal

```python
#unit step signal

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11)

unit_step = np.zeros(n.shape)
unit_step[n >= 0] = 1

plt.step(n, unit_step, where='post', label='Unit Step Signal', linewidth=2)
plt.title('Unit Step Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```

3. Unit Ramp Signal

```python
#unit ramp signal

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 11)

unit_ramp = np.zeros(n.shape)
unit_ramp[n >= 0] = n[n >= 0]

plt.plot(n, unit_ramp, label='Unit Ramp Signal', linewidth=2)
plt.title('Unit Ramp Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(-10, 10)
plt.ylim(-1, 11)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```

4. Exponential Signal

```python
#exponential signal

import numpy as np
import matplotlib.pyplot as plt

# Define time range for exponential signal
t = np.linspace(0, 3, 100)  # from 0 to 3 with 100 points
a = 1  # growth rate

# Create exponential signal
exponential_signal = np.exp(a * t)

# Plot the exponential signal
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st subplot
plt.plot(t, exponential_signal, label='Exponential Signal', color='blue')
plt.title('Exponential Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
```

5. Square Wave Signal

```python
#square wave signal

import numpy as np
import matplotlib.pyplot as plt

from scipy import signal

# Define time range for square wave signal
t_square = np.linspace(0, 3, 1000)  # from 0 to 3 with 1000 points
frequency = 1  # frequency of the square wave

# Create square wave signal
square_wave_signal = signal.square(2 * np.pi * frequency * t_square)

# Plot the square wave signal
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd subplot
plt.plot(t_square, square_wave_signal, label='Square Wave Signal', color='orange')
plt.title('Square Wave Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
```

6. Triangular Signal


```python
#triangular signal

import numpy as np
import matplotlib.pyplot as plt

# Define time range for triangular wave signal
t_triangular = np.linspace(0, 3, 1000)  # from 0 to 3 with 1000 points

# Create triangular wave signal
triangular_signal = signal.sawtooth(2 * np.pi * frequency * t_triangular, 0.5)

# Plot the triangular wave signal
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
plt.plot(t_triangular, triangular_signal, label='Triangular Signal', color='green')
plt.title('Triangular Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# Show all plots
plt.tight_layout()
plt.show()
```
